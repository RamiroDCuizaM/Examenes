## 1. Contexto General

El **Australian Credit Approval Dataset** es un conjunto de datos clásico utilizado ampliamente en aprendizaje automático y estadística. Fue originalmente recopilado para evaluar solicitudes de **tarjetas de crédito** en Australia, y forma parte del repositorio UCI Machine Learning Repository.

### ¿Qué representa?

Cada fila del dataset corresponde a **una solicitud de crédito** realizada por un individuo. El objetivo es determinar si la solicitud debe ser **aprobada (1) o rechazada (0)** en función de una serie de atributos del solicitante.

Los atributos originales han sido **anonimizados** por razones de confidencialidad, por lo que los nombres de las columnas son genéricos (A1–A15). Sin embargo, la comunidad de ML ha documentado el significado aproximado de cada campo basándose en el contexto del dominio crediticio.

### Características clave:
- **690 instancias** (solicitudes de crédito)
- **14 variables predictoras** + **1 variable objetivo**
- **Sin valores faltantes** en esta versión preprocesada
- **Mezcla de variables** continuas y categóricas (todas ya codificadas numéricamente)
- **Problema de clasificación binaria**: aprobado (1) vs. rechazado (0)

Este dataset es ideal para practicar:
- 📊 Análisis Exploratorio de Datos (EDA)
- 🤖 Clasificación binaria (Regresión Logística, SVM, Árboles de Decisión)
- 🧠 Redes Neuronales
- 📉 Regresión (predicción de scores de crédito)
- 🔄 Normalización y preprocesamiento de datos

## 7. Transformaciones a Numérico y Justificación

Este dataset **ya está completamente codificado en valores numéricos**, por lo que no se requiere transformación de texto a número. Sin embargo, realizaremos las siguientes **codificaciones y ajustes explícitos** para mayor claridad y compatibilidad con modelos de ML:

### 🔄 Transformaciones aplicadas:

| Columna | Tipo original | Transformación | Razón |
|---|---|---|---|
| `estado_civil` (A4) | Categórica ordinal (1,2,3) | **One-Hot Encoding** | Evitar relación de orden falsa entre categorías |
| `ocupacion` (A6) | Categórica ordinal (1-9) | **One-Hot Encoding** | Los códigos de ocupación no tienen jerarquía numérica |
| `tipo_cuenta` (A12) | Categórica ordinal (1,2,3) | **One-Hot Encoding** | Igual que estado_civil: categorías sin orden real |
| `genero`, `tiene_tel`, `empleado`, `cred_previo`, `aprobado` | Binarias (0,1) | **Sin cambio** | Ya están correctamente codificadas |
| Variables continuas | Numéricas | **Sin cambio** | Listas para normalización directa |

> 🧠 **¿Por qué One-Hot Encoding?** Las columnas `estado_civil`, `ocupacion` y `tipo_cuenta` contienen categorías sin orden natural. Si las dejamos como números (1, 2, 3), los modelos podrían interpretar que "3 > 2 > 1", lo cual introduce un **sesgo artificial**. Con OHE, cada categoría se convierte en una columna binaria independiente.

```python
# ─────────────────────────────────────────────────────────────
# PASO 1: Copia de trabajo del dataframe original
# ─────────────────────────────────────────────────────────────
df_procesado = df.copy()

print("Dataset ANTES del procesamiento:")
print(f"  Forma: {df_procesado.shape}")
print(f"  Columnas: {list(df_procesado.columns)}")
```

```python
# ─────────────────────────────────────────────────────────────
# PASO 2: One-Hot Encoding para variables categóricas nominales
# ─────────────────────────────────────────────────────────────

cols_categoricas = ['estado_civil', 'ocupacion', 'tipo_cuenta']

print("📊 Distribución de categorías antes del OHE:")
for col in cols_categoricas:
    print(f"\n  {col}: {sorted(df_procesado[col].unique())}")

# Aplicar One-Hot Encoding
df_procesado = pd.get_dummies(
    df_procesado,
    columns=cols_categoricas,
    prefix=['e_civil', 'ocup', 't_cuenta'],
    drop_first=False  # Mantenemos todas las dummies para transparencia
)

# Convertir columnas booleanas a enteros (0/1)
bool_cols = df_procesado.select_dtypes(include='bool').columns
df_procesado[bool_cols] = df_procesado[bool_cols].astype(int)

print("\n✅ One-Hot Encoding aplicado")
print(f"  Nuevas columnas generadas: {list(bool_cols)}")
print(f"  Forma después del OHE: {df_procesado.shape}")
```

```python
# Vista del dataset procesado
print("Dataset DESPUÉS del procesamiento:")
print(f"  Forma: {df_procesado.shape}")
print(f"  Columnas: {list(df_procesado.columns)}")
print()
df_procesado.head()
```

## 8. Separación de Features (X) y Target (y)

```python
# Separar variable objetivo del resto
y = df_procesado['aprobado'].values
X = df_procesado.drop(columns=['aprobado'])

feature_names = list(X.columns)
X = X.values  # Convertir a array numpy

print("📐 Dimensiones:")
print(f"   X (features) : {X.shape}  →  {X.shape[0]} muestras × {X.shape[1]} features")
print(f"   y (target)   : {y.shape}  →  {Counter(y)}")
print()
print("📋 Features en X:")
for i, name in enumerate(feature_names):
    print(f"   [{i:02d}] {name}")
```

## 9. Normalización con `featureNormalize`

La normalización estandariza cada feature para que tenga **media 0** y **desviación estándar 1** (Z-score normalization). Esto es fundamental para:

- 🧠 **Redes neuronales**: evita que gradientes exploten o desaparezcan
- 📏 **SVM y KNN**: los algoritmos basados en distancias son muy sensibles a la escala
- 📉 **Regresión logística**: acelera la convergencia del gradiente
- 🌲 **Árboles de decisión**: en este caso NO es necesario, pero no daña

```python
def featureNormalize(X):
    """
    Normaliza las features usando Z-score (media=0, desv.std=1).
    
    Parámetros:
    -----------
    X : np.ndarray de forma (m, n)
    
    Retorna:
    --------
    X_norm : array normalizado
    mu     : vector de medias (forma: n,)
    sigma  : vector de desviaciones estándar (forma: n,)
    """
    X_norm = X.copy()
    mu = np.zeros(X.shape[1])
    sigma = np.zeros(X.shape[1])

    mu = np.mean(X, axis=0)
    sigma = np.std(X, axis=0)
    X_norm = (X - mu) / sigma

    return X_norm, mu, sigma

# Aplicar normalización
X_norm, mu, sigma = featureNormalize(X)

print("✅ Normalización aplicada")
print(f"   X_norm shape : {X_norm.shape}")
print(f"   Media  post-norm (esperada ~0): {X_norm.mean(axis=0).round(10)}")
print(f"   Std    post-norm (esperada ~1): {X_norm.std(axis=0).round(10)}")
```

```python
# Tabla de mu y sigma por feature
tabla_norm = pd.DataFrame({
    'Feature'   : feature_names,
    'Media (μ)' : mu.round(4),
    'Std  (σ)'  : sigma.round(4),
    'Min original' : X.min(axis=0).round(4),
    'Max original' : X.max(axis=0).round(4)
})

print("📊 Parámetros de Normalización por Feature:")
tabla_norm
```

```python
# Visualización: Antes vs Después de normalizar (variables continuas)
vars_idx = [feature_names.index(c) for c in ['edad', 'deuda', 'anios_empleo', 'ingreso']]
var_labels = ['edad', 'deuda', 'anios_empleo', 'ingreso']

fig, axes = plt.subplots(2, 4, figsize=(16, 7))

for i, (idx, label) in enumerate(zip(vars_idx, var_labels)):
    # Antes
    axes[0, i].hist(X[:, idx], bins=30, color='#e67e22', edgecolor='white', alpha=0.85)
    axes[0, i].set_title(f'{label}\n(Original)', fontweight='bold')
    axes[0, i].set_ylabel('Frecuencia' if i == 0 else '')

    # Después
    axes[1, i].hist(X_norm[:, idx], bins=30, color='#27ae60', edgecolor='white', alpha=0.85)
    axes[1, i].set_title(f'{label}\n(Normalizada)', fontweight='bold')
    axes[1, i].set_ylabel('Frecuencia' if i == 0 else '')

plt.suptitle('Distribuciones: Antes vs Después de Normalizar', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 10. División en Conjuntos de Entrenamiento y Prueba

```python
from sklearn.model_selection import train_test_split

# División 80/20, estratificada para mantener proporción de clases
X_train, X_test, y_train, y_test = train_test_split(
    X_norm, y,
    test_size=0.2,
    random_state=42,
    stratify=y  # Mantiene la proporción de clases en cada subset
)

print("✅ División completada (80% train / 20% test)")
print(f"   X_train : {X_train.shape}  |  y_train: {Counter(y_train)}")
print(f"   X_test  : {X_test.shape}   |  y_test : {Counter(y_test)}")
print()
print(f"   Proporción aprobados en train: {y_train.mean():.2%}")
print(f"   Proporción aprobados en test : {y_test.mean():.2%}")
```
