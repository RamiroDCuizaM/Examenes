# Dataset Ames Housing (Ames, Iowa)

## Contexto del dataset

Este conjunto de datos recoge **ventas de viviendas residenciales en Ames (Iowa, EE. UU.)** entre **2006 y 2010**. Es una versión ampliada del dataset publicado por **De Cock (2011)** en el *Journal of Statistics Education*, muy utilizado en enseñanza de regresión y machine learning porque combina **variables numéricas, categóricas y ordinales** con un número manejable de observaciones.

**Objetivos típicos de modelado:**

- **Regresión:** predecir `SalePrice` a partir de las características de la vivienda y del lote.
- **Clasificación:** definir una variable objetivo categórica (por ejemplo, rangos de precio, tipo de barrio codificado en clases, o venta "cara" vs "barata").
- **Redes neuronales:** usar la matriz de características numéricas (normalizada) y el vector objetivo continuo o one-hot para clases.

**Tamaño habitual:** del orden de **~2930 filas** y **~80 variables** más identificadores, según la versión del archivo.

En la siguiente tabla se describe **cada columna** según la documentación habitual del dataset Ames Housing.

## 2. Plan de preparación para machine learning

Pasos que aplicaremos y **por qué**:

1. **Eliminar `Order` y `PID`:** son identificadores; no aportan señal para generalizar y pueden introducir ruido o fugas si se interpretaran como orden temporal.
2. **Tratar `MS SubClass` como categórico:** los códigos son etiquetas de tipo de vivienda, no una escala continua; se convertirá a tipo texto y luego a columnas dummy numéricas.
3. **Valores faltantes en variables categóricas:** en Ames, un ausente suele significar **"no aplica"** (ej. sin piscina, sin alley). Los unificamos a la cadena `"None"` para poder codificarlos explícitamente.
4. **Variables ordinales de calidad** (`Exter Qual`, `Kitchen Qual`, etc.): tienen orden natural (Ex > Gd > TA > …). Las mapeamos a **enteros 0,1,…** para conservar el orden (mejor que one-hot si solo usamos una dimensión por variable).
5. **Variables numéricas con NaN** (ej. `Lot Frontage`, `Garage Yr Blt`): imputamos con la **mediana** por robustez a outliers; para `Mas Vnr Area` el 0 también es razonable cuando no hay mampostería tras imputar tipo.
6. **Variables nominales restantes:** **one-hot encoding** (`OneHotEncoder` de scikit-learn) para convertir cada categoría en columnas 0/1. `handle_unknown='ignore'` evita fallos si en datos nuevos aparece una categoría no vista en entrenamiento.
7. **Variable objetivo:** `SalePrice` se deja como vector `y` para regresión. Para **clasificación** puedes crear bins (ej. cuartiles) a partir de `y`.
8. **Escalado (opcional):** para redes neuronales y muchos algoritmos basados en distancias, estandarizar (`StandardScaler`) las características numéricas finales.

---

### Resumen de transformaciones y significado

| Transformación | Qué hace | Para qué sirve |
|----------------|----------|----------------|
| Drop IDs | Quita columnas de solo identificación | Evita ruido y uso incorrecto como predictor |
| `astype(str)` en `MS SubClass` + dummies | Categorías → columnas binarias | Modelos solo numéricos |
| `fillna('None')` en texto | Ausente = categoría explícita | Evita perder filas y refleja "no existe" |
| Mapa ordinal calidad | Ex→5, …, NA→0 | Orden en un solo número por variable |
| `SimpleImputer` mediana | Rellena huecos en numéricos | Completa datos para sklearn/keras |
| `OneHotEncoder` | Nominal → 0/1 | Lineal, SVM, redes, etc. |
| `StandardScaler` (opcional) | Media 0, varianza 1 | Convergencia en redes y kNN |


```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

df_ml = df.copy()
df_ml = df_ml.drop(columns=["Order", "PID"], errors="ignore")

# MS SubClass como categoría (códigos de tipo de edificación)
df_ml["MS SubClass"] = df_ml["MS SubClass"].astype(str)

# Columnas con escala ordinal de calidad (Ames Housing)
ordinal_quality_cols = [
    "Exter Qual", "Exter Cond", "Bsmt Qual", "Bsmt Cond", "Heating QC",
    "Kitchen Qual", "Fireplace Qu", "Garage Qual", "Garage Cond", "Pool QC",
]
quality_map = {"Ex": 5, "Gd": 4, "TA": 3, "Fa": 2, "Po": 1, "NA": 0, "None": 0}

for col in ordinal_quality_cols:
    if col in df_ml.columns:
        df_ml[col] = df_ml[col].astype(str).replace("nan", "NA")
        df_ml[col] = df_ml[col].map(quality_map).fillna(0)

# Bsmt Exposure: orden aproximado (más exposición / mejor → mayor número)
bsmt_exp_map = {"Gd": 4, "Av": 3, "Mn": 2, "No": 1, "NA": 0, "None": 0, "nan": 0}
if "Bsmt Exposure" in df_ml.columns:
    df_ml["Bsmt Exposure"] = df_ml["Bsmt Exposure"].astype(str).replace("nan", "NA")
    df_ml["Bsmt Exposure"] = df_ml["Bsmt Exposure"].map(bsmt_exp_map).fillna(0)

# BsmtFin Type: orden de "terminado" típico en documentación (GLQ mejor que Unf para calidad de espacio)
bsmt_fin_map = {
    "GLQ": 6, "ALQ": 5, "BLQ": 4, "Rec": 3, "LwQ": 2, "Unf": 1, "NA": 0, "None": 0, "nan": 0,
}
for col in ["BsmtFin Type 1", "BsmtFin Type 2"]:
    if col in df_ml.columns:
        df_ml[col] = df_ml[col].astype(str).replace("nan", "NA")
        df_ml[col] = df_ml[col].map(bsmt_fin_map).fillna(0)

# Garage Finish
gar_fin_map = {"Fin": 3, "RFn": 2, "Unf": 1, "NA": 0, "None": 0, "nan": 0}
if "Garage Finish" in df_ml.columns:
    df_ml["Garage Finish"] = df_ml["Garage Finish"].astype(str).replace("nan", "NA")
    df_ml["Garage Finish"] = df_ml["Garage Finish"].map(gar_fin_map).fillna(0)

# Functional (parcialmente ordinal)
func_map = {
    "Typ": 8, "Min1": 7, "Min2": 6, "Mod": 5, "Maj1": 4, "Maj2": 3, "Sev": 2, "Sal": 1,
}
if "Functional" in df_ml.columns:
    df_ml["Functional"] = df_ml["Functional"].astype(str)
    df_ml["Functional"] = df_ml["Functional"].map(func_map).fillna(8)

print("Ordinales recodificados.")
```

```python
# Separar objetivo
y = df_ml["SalePrice"].astype(float)
X = df_ml.drop(columns=["SalePrice"])

# Identificar columnas: numéricas vs categóricas (object/string)
num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = X.select_dtypes(include=["object", "string"]).columns.tolist()

# Rellenar categorías faltantes como 'None'
for c in cat_cols:
    X[c] = X[c].astype(str).replace("nan", "None")
    X[c] = X[c].fillna("None")

numeric_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler()),
    ]
)

categorical_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False)),
    ]
)

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, num_cols),
        ("cat", categorical_transformer, cat_cols),
    ],
)

X_processed = preprocess.fit_transform(X)

# Nombres de columnas después de one-hot
ohe = preprocess.named_transformers_["cat"].named_steps["onehot"]
cat_feature_names = list(ohe.get_feature_names_out(cat_cols))
feature_names = num_cols + cat_feature_names

X_df = pd.DataFrame(X_processed, columns=feature_names, index=X.index)

print("Matriz de características:", X_df.shape)
print("Vector objetivo y:", y.shape)
X_df.head()
```

### Qué acaba de ocurrir en el bloque anterior (detalle)

- **`num_cols`:** todas las columnas ya numéricas (incluye `Overall Qual`, años, superficies, etc.). Sobre ellas: **imputación por mediana** (por si quedó algún NaN) y **`StandardScaler`** para media 0 y desviación 1 → recomendable para **redes neuronales**, SVM o kNN.
- **`cat_cols`:** textos restantes (`Neighborhood`, `MS Zoning`, …). Tras `fillna('None')`, **`OneHotEncoder`** crea una columna binaria por categoría (sparse=False para obtener array denso). `handle_unknown='ignore'` evita errores si en test aparece una categoría nueva.
- **Resultado `X_df`:** solo números (`float`), listo para `sklearn`, **TensorFlow/Keras** o **PyTorch**.


```python
# Ejemplo: partición entrenamiento / prueba y baseline de regresión
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, r2_score

X_train, X_test, y_train, y_test = train_test_split(
    X_df.values, y.values, test_size=0.2, random_state=42
)

model = Ridge(alpha=1.0, random_state=42)
model.fit(X_train, y_train)
pred = model.predict(X_test)
print("MAE:", mean_absolute_error(y_test, pred))
print("R2:", r2_score(y_test, pred))
```
